```python
import os
import pandas as pd
import numpy as np
import dask.dataframe as dd
from numpy import dot
from numpy.linalg import norm
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import scipy.stats as ss
import json
import pathlib
from matplotlib_venn import venn2,venn3,venn2_circles,venn3_circles
from matplotlib_venn.layout.venn2 import DefaultLayoutAlgorithm

%matplotlib inline  
```

```python
# Set data input folder
# Download files described in README.md to this folder
input_folder = "inputs"

# Set output folder, subfolder
output_folder = "outputs"
if not os.path.exists(output_folder):
    os.makedirs(output_folder, exist_ok=True)
if not os.path.exists(os.path.join(output_folder,"figure_panels")):
    os.makedirs(os.path.join(output_folder,"figure_panels"), exist_ok=True)
```

```python
# Load hits from the hit calling process
whole_cell_hits_cp = pd.read_csv('../CP_1_FDR/outputs/HeLa_CP_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
whole_cell_hits_cp_list = list(whole_cell_hits_cp.Gene)
comp_spec_hits_cp = pd.read_csv('../CP_1_FDR/outputs/HeLa_CP_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')
comp_spec_hits_cp_list = list(comp_spec_hits_cp.Gene)
all_hits_cp = whole_cell_hits_cp_list + comp_spec_hits_cp_list

# Load hits from the hit calling process for SABER-1-2
whole_cell_hits_saber_1_2 = pd.read_csv('../SABER-1-2/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
whole_cell_hits_saber_1_2_list = list(whole_cell_hits_saber_1_2.Gene)
comp_spec_hits_saber_1_2 = pd.read_csv('../SABER-1-2/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')
comp_spec_hits_saber_1_2_list = list(comp_spec_hits_saber_1_2.Gene)
all_hits_saber_1_2 = whole_cell_hits_saber_1_2_list + comp_spec_hits_saber_1_2_list

# Load hits from the hit calling process for SABER
whole_cell_hits_saber = pd.read_csv('../SABER/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
whole_cell_hits_saber_list = list(whole_cell_hits_saber.Gene)
comp_spec_hits_saber = pd.read_csv('../SABER/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')
comp_spec_hits_saber_list = list(comp_spec_hits_saber.Gene)
all_hits_saber = whole_cell_hits_saber_list + comp_spec_hits_saber_list

# Load hits from the hit calling process for SABER-1-2 without WGA or ConA features
whole_cell_hits_no_wga_cona = pd.read_csv('../SABER-1-2_no_WGA_no_ConA/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
comp_spec_hits_no_wga_cona = pd.read_csv('../SABER-1-2_no_WGA_no_ConA/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')

# Load hits from the hit calling process for SABER-1-2 without Calnexin or GM130 features
whole_cell_hits_no_cal_gm130 = pd.read_csv('../SABER-1-2-no_Calnexin_no_GM130/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
comp_spec_hits_no_cal_gm130 = pd.read_csv('../SABER-1-2-no_Calnexin_no_GM130/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')
```

```python
# Create lists for common and specific genes for CP and SABER arm
cp_saber_common = [gene for gene in all_hits_cp if gene in all_hits_saber]
cp_saber_saber_only  = [gene for gene in all_hits_saber if gene not in all_hits_cp]
cp_saber_cp_only = [gene for gene in all_hits_cp if gene not in all_hits_saber]

print('SABER arm hits =',len(all_hits_saber),'CP arm hits =',len(all_hits_cp))
print('Common hits between arms =',len(cp_saber_common))
print('SABER only hits =',len(cp_saber_saber_only))
print('CP only hits =',len(cp_saber_cp_only))
```

```python
#plt.figure(figsize=(5,10))
fig, ax = plt.subplots(figsize=(5,10))

norm_factor = (len(cp_saber_saber_only)+len(cp_saber_cp_only)+len(cp_saber_common))/2
reg_sizes = (len(cp_saber_saber_only)+norm_factor, len(cp_saber_cp_only)+norm_factor, len(cp_saber_common)+norm_factor)

v = venn2(
    subsets = (len(cp_saber_saber_only), len(cp_saber_cp_only), len(cp_saber_common)), 
    set_labels = ('SABER', 'CP'),
    layout_algorithm=DefaultLayoutAlgorithm(fixed_subset_sizes=reg_sizes))

c = venn2_circles(
    subsets = (len(cp_saber_saber_only), len(cp_saber_cp_only), len(cp_saber_common)), 
    layout_algorithm=DefaultLayoutAlgorithm(fixed_subset_sizes=reg_sizes),
    linestyle='dashed', 
    linewidth=1, 
    color="grey")

lbl = v.get_label_by_id('B')
x, y = lbl.get_position()
lbl.set_position((x+0.1, y-0.03))  

fig.savefig(os.path.join(output_folder,'figure_panels','CP498_CP_SABER_hits_venn_diagram.png'),dpi = 300,bbox_inches='tight')
plt.show()
```

```python
# Create lists for common and specific genes for CP and SABER (Plate 1&2) arm
cp_saber_1_2_common = [gene for gene in all_hits_cp if gene in all_hits_saber_1_2]
cp_saber_1_2_saber_1_2_only  = [gene for gene in all_hits_saber_1_2 if gene not in all_hits_cp]
cp_saber_1_2_cp_only = [gene for gene in all_hits_cp if gene not in all_hits_saber_1_2]

print('SABER-1-2 arm hits =',len(all_hits_saber_1_2),'CP arm hits =',len(all_hits_cp))
print('Common hits between arms =',len(cp_saber_1_2_common))
print('SABER-1-2 only hits =',len(cp_saber_1_2_saber_1_2_only))
print('CP only hits =',len(cp_saber_1_2_cp_only))
```

```python
fig, ax = plt.subplots(figsize=(5,10))

norm_factor = (len(cp_saber_1_2_saber_1_2_only)+len(cp_saber_1_2_cp_only)+len(cp_saber_1_2_common))/2
reg_sizes = (len(cp_saber_1_2_saber_1_2_only)+norm_factor, len(cp_saber_1_2_cp_only)+norm_factor, len(cp_saber_1_2_common)+norm_factor)

v = venn2(
    subsets = (len(cp_saber_1_2_saber_1_2_only), len(cp_saber_1_2_cp_only), len(cp_saber_1_2_common)), 
    set_labels = ('SABER(Plates 1&2)', 'CP'),
    layout_algorithm=DefaultLayoutAlgorithm(fixed_subset_sizes=reg_sizes))

c = venn2_circles(
    subsets = (len(cp_saber_1_2_saber_1_2_only), len(cp_saber_1_2_cp_only), len(cp_saber_1_2_common)), 
    layout_algorithm=DefaultLayoutAlgorithm(fixed_subset_sizes=reg_sizes),
    linestyle='dashed', 
    linewidth=1, 
    color="grey")

lbl = v.get_label_by_id('A')
x, y = lbl.get_position()
lbl.set_position((x+0.25, y))

lbl = v.get_label_by_id('B')
x, y = lbl.get_position()
lbl.set_position((x+0.1, y-0.02))  

fig.savefig(os.path.join(output_folder,'figure_panels','CP498_CP_SABER_1_2_hits_venn_diagram.png'),dpi = 300,bbox_inches='tight')
plt.show()
```
