```python
import os
import pandas as pd
import numpy as np
import dask.dataframe as dd
from numpy import dot
from numpy.linalg import norm
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import scipy.stats as ss
import json
import pathlib

%matplotlib inline  
```

```python
# Set data input folder
# Download files described in README.md to this folder
input_folder = "inputs"

# Set output folder, subfolder
output_folder = "outputs"
if not os.path.exists(output_folder):
    os.makedirs(output_folder, exist_ok=True)
if not os.path.exists(os.path.join(output_folder,"figure_panels")):
    os.makedirs(os.path.join(output_folder,"figure_panels"), exist_ok=True)
```

```python
# Load hits from the hit calling process
whole_cell_hits_cp = pd.read_csv('../CP_1_FDR/outputs/HeLa_CP_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
comp_spec_hits_cp = pd.read_csv('../CP_1_FDR/outputs/HeLa_CP_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')

# Load hits from the hit calling process for SABER-1-2
whole_cell_hits_saber_1_2 = pd.read_csv('../SABER-1-2/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
comp_spec_hits_saber_1_2 = pd.read_csv('../SABER-1-2/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')

# Load hits from the hit calling process for SABER
whole_cell_hits_saber = pd.read_csv('../SABER/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
comp_spec_hits_saber = pd.read_csv('../SABER/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')

# Load hits from the hit calling process for SABER-1-2 without WGA or ConA features
whole_cell_hits_no_wga_cona = pd.read_csv('../SABER-1-2_no_WGA_no_ConA/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
comp_spec_hits_no_wga_cona = pd.read_csv('../SABER-1-2_no_WGA_no_ConA/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')

# Load hits from the hit calling process for SABER-1-2 without Calnexin or GM130 features
whole_cell_hits_no_cal_gm130 = pd.read_csv('../SABER-1-2-no_Calnexin_no_GM130/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_whole_cell_hits.csv')
comp_spec_hits_no_cal_gm130 = pd.read_csv('../SABER-1-2-no_Calnexin_no_GM130/outputs/HeLa_SABER_plate_level_median_per_feat_sig_genes_1_FDR_compartment_specific_hits.csv')
```

```python
mpl.rc('axes', linewidth=0.7)
mpl.rc('xtick', labelsize=14)
mpl.rc('xtick', labelsize=14)

fig, ax = plt.subplots(figsize=(7,5))
colors = sns.color_palette('GnBu')[:8]
hits_plot = pd.DataFrame({'Compartment': [len(comp_spec_hits_cp), len(comp_spec_hits_saber_1_2), len(comp_spec_hits_saber)],
                         'Whole cell': [len(whole_cell_hits_cp),len(whole_cell_hits_saber_1_2), len(whole_cell_hits_saber)]},
                        index=['CP','SABER(Plates 1&2)','SABER'])

hits_plot.plot(kind='bar', 
               stacked=True,
               color=[colors[1], colors[4]],
               legend = 'reverse',
               rot = 0,
               fontsize= 14,
               width = 0.5,
               ax=ax)

ax.set_title('Genes with significant signal (1% FDR)',size=14) #Genes with significant signal above noise
ax.set_ylabel('Number of genes',size=14)

ax.bar_label(ax.containers[0],label_type='center',size=13)
ax.bar_label(ax.containers[1],label_type='center',size=13)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(loc="upper left", bbox_to_anchor=(0.9,1),frameon=False, ncol=1, fontsize =14)
fig.savefig(os.path.join(output_folder,'figure_panels','CP498_HeLa_CP_SABER-1-2_SABER_whole_cell_compartment_hits.png'),dpi = 300,bbox_inches='tight')
```

```python
mpl.rc('axes', linewidth=0.7)
mpl.rc('xtick', labelsize=14)
mpl.rc('xtick', labelsize=14)

fig, ax = plt.subplots(figsize=(7,5))
colors = sns.color_palette('GnBu')[:8]
hits_plot = pd.DataFrame({'Compartment': [len(comp_spec_hits_saber_1_2), len(comp_spec_hits_no_wga_cona), len(comp_spec_hits_no_cal_gm130)],
                         'Whole cell': [len(whole_cell_hits_saber_1_2),len(whole_cell_hits_no_wga_cona), len(whole_cell_hits_no_cal_gm130)]},
                        index=['SABER','No WGA/ConA','No Calnexin/GM130'])

hits_plot.plot(kind='bar', 
               stacked=True,
               color=[colors[1], colors[4]],
               legend = 'reverse',
               rot = 0,
               fontsize= 14,
               width = 0.5,
               ax=ax)

ax.set_title('Genes with significant signal - plates 1&2 (1% FDR)',size=14) #Genes with significant signal above noise
ax.set_ylabel('Number of genes',size=14)

ax.bar_label(ax.containers[0],label_type='center',size=13)
ax.bar_label(ax.containers[1],label_type='center',size=13)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(loc="upper left", bbox_to_anchor=(0.9,1),frameon=False, ncol=1, fontsize =14)
fig.savefig(os.path.join(output_folder,'figure_panels','CP498_HeLa_SABER-1-2_no_wga_cona_caln_gma130_whole_cell_compartment_hits.png'),dpi = 300,bbox_inches='tight')
```
