```python
import os
import pandas as pd
import numpy as np
import dask.dataframe as dd
from numpy import dot
from numpy.linalg import norm
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import scipy.stats as ss
import json
import pathlib
from pycytominer import aggregate, normalize

%matplotlib inline  
```

```python
# Set data input folder
# Download files described in README.md to this folder
input_folder = "inputs"

# Set output folder, subfolder
output_folder = "outputs"
if not os.path.exists(output_folder):
    os.makedirs(output_folder, exist_ok=True)
if not os.path.exists(os.path.join(output_folder,"figure_panels")):
    os.makedirs(os.path.join(output_folder,"figure_panels"), exist_ok=True)
```

```python
# Load the subsampled, aggregated profiles from the previous step if you are starting from here
# These files are not stored in the Github repository - see the README for download instructions

subsample_ns = [100,200,300,400,500,750,1000]
subsample_dfs_dictionary_cp = {} 
for n in subsample_ns:
    aggregate_df_n = pd.read_csv(f'../CP_5_FDR/single_cell_profiles/outputs/CP498_cp_arm_subsampled_aggregated_{n}_cells.csv')
    subsample_dfs_dictionary_cp[n] = aggregate_df_n
```

```python
# Load the subsampled, aggregated profiles from the previous step if you are starting from here
# These files are not stored in the Github repository - see the README for download instructions
subsample_dfs_dictionary_saber_1_2 = {} 
for n in subsample_ns:
    aggregate_df_n = pd.read_csv(f'../SABER-1-2/single_cell_profiles/CP498_SABER_arm_1_2_subsampled_aggregated_{n}_cells.csv')
    subsample_dfs_dictionary_saber_1_2[n] = aggregate_df_n
```

```python
# Load the subsampled, aggregated profiles from the previous step if you are starting from here
# These files are not stored in the Github repository - see the README for download instructions
subsample_dfs_dictionary_saber = {} 
for n in subsample_ns:
    aggregate_df_n = pd.read_csv(f'../SABER/single_cell_profiles/CP498_SABER_arm_subsampled_aggregated_{n}_cells.csv')
    subsample_dfs_dictionary_saber[n] = aggregate_df_n
```

```python
def cosine_to_df(df_temp, cosine_array, i):
    cosine_list = cosine_array[i]
    gene_list = list(df_temp.index)
    cosine_df = pd.DataFrame(index=gene_list)
    cosine_df['cosine'] = cosine_list
    cosine_df = cosine_df.sort_values('cosine',ascending=False)   
    return cosine_df

def ap_from_cosine_df(cosine_df,gene,n=10):    
    #print(cosine_df.iloc[:20])
    index_list = list(cosine_df.index)
    boolean = [1 if  i == gene else 0 for i in index_list ]
    grades_list=[]
    for i in range(2,n+2):
        pre_grade = sum(boolean[1:i])/(i-1)
        grades_list.append(pre_grade*boolean[i-1])
    return sum(grades_list)/3

def calculate_map(df_guide, gene):
    df_temp = df_guide.query("Metadata_Foci_Barcode_MatchedTo_GeneCode == 'nontargeting' | Metadata_Foci_Barcode_MatchedTo_GeneCode == @gene")
    df_temp = df_temp.drop(['Metadata_Foci_Barcode_MatchedTo_Barcode'],axis=1)
    df_temp = df_temp.set_index("Metadata_Foci_Barcode_MatchedTo_GeneCode")
    #print(df_temp)
    ap_list = []
    cosine_array = cosine_similarity(df_temp)
    for guide in range(4):
        cosine_df = cosine_to_df(df_temp, cosine_array, guide)
        #print(cosine_df[:10])
        guide_ap = ap_from_cosine_df(cosine_df,gene,10)
        ap_list.append(guide_ap)
    return np.mean(ap_list)
```

# Calculate the subsampling between CP, SABER and SABER-1-2 for all available genes for each group

```python
# calculate the mAP values for cp arm
rep_results_cp = {}
for n in subsample_ns:
    genes_list = list(subsample_dfs_dictionary_cp[n].Metadata_Foci_Barcode_MatchedTo_GeneCode.unique())
    map_list = []
    for i in range(len(genes_list)):
        gene = genes_list[i]
        gene_map = calculate_map(subsample_dfs_dictionary_cp[n], gene)
        map_list.append([gene, gene_map])
    rep_results_cp[n] = map_list

# calculate the mAP values for SABER plates 1 & 2
rep_results_saber_1_2 = {}
for n in subsample_ns:
    genes_list = list(subsample_dfs_dictionary_saber_1_2[n].Metadata_Foci_Barcode_MatchedTo_GeneCode.unique())
    map_list = []
    for i in range(len(genes_list)):
        gene = genes_list[i]
        gene_map = calculate_map(subsample_dfs_dictionary_saber_1_2[n], gene)
        map_list.append([gene, gene_map])
    rep_results_saber_1_2[n] = map_list

# calculate the mAP values FOR SABER arm
rep_results_saber = {}
for n in subsample_ns:
    genes_list = list(subsample_dfs_dictionary_saber[n].Metadata_Foci_Barcode_MatchedTo_GeneCode.unique())
    map_list = []
    for i in range(len(genes_list)):
        gene = genes_list[i]
        gene_map = calculate_map(subsample_dfs_dictionary_saber[n], gene)
        map_list.append([gene, gene_map])
    rep_results_saber[n] = map_list
```

```python
ys_cp = []
for n in subsample_ns:
    y = [i[1] for i in rep_results_cp[n] if i[1]]
    ys_cp.append(np.mean(y))

ys_saber_1_2 = []
for n in subsample_ns:
    y = [i[1] for i in rep_results_saber_1_2[n] if i[1]]
    ys_saber_1_2.append(np.mean(y))

ys_saber = []
for n in subsample_ns:
    y = [i[1] for i in rep_results_saber[n] if i[1]]
    ys_saber.append(np.mean(y))

mpl.rc('axes', linewidth=0.7)
mpl.rc('ytick', labelsize=12)
mpl.rc('xtick', labelsize=12)

fig, ax = plt.subplots(figsize=(7,5))

ax = sns.lineplot(x=subsample_ns, 
                  y=ys_cp, 
                  errorbar=None, 
                  color='tab:red',
                  marker='o',
                  markersize=6,
                  linestyle=':',
                  linewidth= 1.5,
                  legend=True,
                  label='CP'
                 )

ax = sns.lineplot(x=subsample_ns, 
                  y=ys_saber_1_2, 
                  errorbar=None, 
                  color='tab:green',
                  marker='o',
                  markersize=6,
                  linestyle=':',
                  linewidth= 1.5,
                  legend=True,
                  label='SABER Plates 1&2'
                 )

ax = sns.lineplot(x=subsample_ns, 
                  y=ys_saber, 
                  errorbar='sd', 
                  color='tab:blue',
                  marker='o',
                  markersize=6,
                  linestyle=':',
                  linewidth= 1.5,
                  legend=True,
                  label='SABER'
                 )

ax.set_title('',size=14)
ax.set_xlabel('Guide level representation',size=12)
ax.set_ylabel('Average mAP',size=12)
ax.set_xticks(range(0,1001,100))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig.savefig(os.path.join(output_folder,'figure_panels','CP498_subsample_average_mAP_CP_SABER_1_2_SABER_all_genes.png'),dpi = 300,bbox_inches='tight')
plt.show()
```

# Calculate the subsampling between CP, SABER and SABER-1-2 for **common** genes

```python
genes_list_cp = list(subsample_dfs_dictionary_cp[n].Metadata_Foci_Barcode_MatchedTo_GeneCode.unique())
genes_list_saber_1_2 = list(subsample_dfs_dictionary_saber_1_2[n].Metadata_Foci_Barcode_MatchedTo_GeneCode.unique())
genes_list_saber = list(subsample_dfs_dictionary_saber[n].Metadata_Foci_Barcode_MatchedTo_GeneCode.unique())

def find_common(list1, list2, list3):
    common = set()
    for elem in list1:
        if elem in list2 and elem in list3:
            common.add(elem)
    return common 
common_genes = list(find_common(genes_list_cp, genes_list_saber_1_2, genes_list_saber))
print(len(common_genes))
```

```python
df_results_cp = pd.DataFrame(index=list(np.arange(0,len(common_genes)*7)),columns=['gene','subsample','mAP'])
counter = 0
for n in subsample_ns:
    for i in range(len(common_genes)):
        gene = common_genes[i]
        gene_map = calculate_map(subsample_dfs_dictionary_cp[n], gene)
        df_results_cp.loc[counter,'gene'] = gene
        df_results_cp.loc[counter,'subsample'] = n
        df_results_cp.loc[counter,'mAP'] = gene_map
        counter+=1

df_results_saber_1_2 = pd.DataFrame(index=list(np.arange(0,len(common_genes)*7)),columns=['gene','subsample','mAP'])
counter = 0
for n in subsample_ns:
    for i in range(len(common_genes)):
        gene = common_genes[i]
        gene_map = calculate_map(subsample_dfs_dictionary_saber_1_2[n], gene)
        df_results_saber_1_2.loc[counter,'gene'] = gene
        df_results_saber_1_2.loc[counter,'subsample'] = n
        df_results_saber_1_2.loc[counter,'mAP'] = gene_map
        counter+=1

df_results_saber = pd.DataFrame(index=list(np.arange(0,len(common_genes)*7)),columns=['gene','subsample','mAP'])
counter = 0
for n in subsample_ns:
    for i in range(len(common_genes)):
        gene = common_genes[i]
        gene_map = calculate_map(subsample_dfs_dictionary_saber[n], gene)
        df_results_saber.loc[counter,'gene'] = gene
        df_results_saber.loc[counter,'subsample'] = n
        df_results_saber.loc[counter,'mAP'] = gene_map
        counter+=1
```

```python
fig, ax = plt.subplots(figsize=(7,5))
errorbar='se'
ax = sns.lineplot(data=df_results_cp,
                  x='subsample', 
                  y='mAP', 
                  errorbar=errorbar,
                  color='tab:red',
                  marker='o',
                  markersize=6,
                  linestyle=':',
                  linewidth= 1.5,
                  legend=True,
                  label='CP'
                 )

ax = sns.lineplot(data=df_results_saber_1_2,
                  x='subsample', 
                  y='mAP', 
                  errorbar=errorbar,
                  color='tab:green',
                  marker='o',
                  markersize=6,
                  linestyle=':',
                  linewidth= 1.5,
                  legend=True,
                  label='SABER Plates 1&2'
                 )

ax = sns.lineplot(data=df_results_saber,
                  x='subsample', 
                  y='mAP', 
                  errorbar=errorbar,
                  color='tab:blue',
                  marker='o',
                  markersize=6,
                  linestyle=':',
                  linewidth= 1.5,
                  legend=True,
                  label='SABER'
                 )


ax.set_title('',size=14)
ax.set_xlabel('Guide level representation',size=12)
ax.set_ylabel('Average mAP',size=12)
ax.set_xticks(range(0,1001,100))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig.savefig(os.path.join(output_folder,'figure_panels','CP498_subsample_average_mAP_CP_SABER_1_2_SABER_common_genes.png'),dpi = 300,bbox_inches='tight')
plt.show()
```
